```python
import pandas as pd
from datetime import datetime, timedelta

now = datetime.now()
df = pd.read_excel("datasets/generated_flights.xlsx")
```

```python
df['sta-std'] = df['sta'] - df['std']
df['std-now'] = df['std'] - now
df['bool(std-now)<1'] = df['std-now'] < timedelta(days = 1)
df['bool(sta-std)<2'] = df['sta-std'] < timedelta(days = 2)
df['bool(sta-std)>1'] = df['sta-std'] > timedelta(days = 1)
df['2>bool(sta-std)>1'] = df['bool(sta-std)<2'] * df['bool(sta-std)>1']
df.drop(columns=['bool(sta-std)>1','bool(sta-std)<2'])
```

```python
class_df = pd.read_excel("datasets/class_weight.xlsx")
```

```python
class_df
```

```python
df['free_weight'] = df['f_availability'] * class_df.loc[class_df['class'] == 'f'].loc[0, 'weight/kg']
```

```python
df
```

```python
flight_output = df.sort_values(by='free_weight', ascending=False, ignore_index=True).head(1)
```

```python
product_df = pd.read_excel('datasets/Food_and_Wines.xlsx')
```

```python
df_output = product_df.loc[product_df['countries'] == flight_output.loc[0,'from']]
```
